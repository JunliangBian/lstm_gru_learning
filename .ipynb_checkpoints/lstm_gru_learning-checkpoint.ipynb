```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from tqdm import tqdm
import random
from torchvision import datasets, transforms
import math

# 设置随机种子
def set_seed(seed=125):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(125)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
transform = transforms.ToTensor()
train_dataset = datasets.MNIST(root='./data', 
                            train=True, 
                            transform=transform,
                            download=True) #MNIST 训练数据集60,000 张手写数字图片
 
test_dataset = datasets.MNIST(root='./data', 
                           train=False, 
                           transform=transform)
 
batch_size = 100
n_iters = 6000 #训练的总迭代次数
# n_iters / (len(train_dataset) / batch_size) 计算总 epoch 数（6000 / 600 = 10）。
num_epochs = n_iters / (len(train_dataset) / batch_size) # len(train_dataset) / batch_size 计算每个 epoch 中的迭代次数（60,000 / 100 = 600 次）。
num_epochs = int(num_epochs)
```

```python
train_loader = DataLoader(dataset=train_dataset, 
                           batch_size=batch_size, 
                           shuffle=True)
 
test_loader = DataLoader(dataset=test_dataset, 
                          batch_size=batch_size, 
                          shuffle=False)
```

## GRU cell implmentation

### 单个时间步计算单元

```python
class GRUCell(nn.Module):
    def __init__(self, input_size, hidden_size, bias=True):
        super(GRUCell, self).__init__()
        self.input_size = input_size # 输入维度
        self.hidden_size = hidden_size # 隐藏层维度
        self.bias = bias # 偏置
        self.x2h = nn.Linear(input_size,3*hidden_size,bias = bias)
        self.h2h = nn.Linear(hidden_size,3*hidden_size,bias = bias)
        self.reset_parameters()
        
    def reset_parameters(self):
        std = 1.0 / math.sqrt(self.hidden_size)
        for w in self.parameters(): # nn.Module 的内置方法
            w.data.uniform_(-std,std)
        
    def forward(self,x,hidden):
        x = x.view(-1,x.size(1)) # 将输入张量 x 重新调整为二维形状
        
        # 线性变换
        gate_x = self.x2h(x) # [batch_size, 3 * hidden_size]
        gate_h = self.h2h(hidden) # [batch_size, 3 * hidden_size]
        
        # 去除可能的额外维度（兼容性处理）
        gate_x = gate_x.squeeze() # 移除张量 gate_x 中所有大小为1的维度。
        gate_h = gate_h.squeeze()
            
        # 将线性变换结果拆分为3个门，沿着第1个维度（列）进行分割。
        i_r, i_u, i_n = gate_x.chunk(3, 1)  # 输入相关的门
        h_r, h_u, h_n = gate_h.chunk(3, 1)  # 隐藏状态相关的门
        
        resetgate = torch.sigmoid(i_r + h_r) # 遗忘门
        updategate =  torch.sigmoid(i_u + h_u) # 更新门
        hiddengate = torch.tanh(i_n + (resetgate * h_n)) #候选隐藏状态
        
        hy = hiddengate + updategate * (hidden - hiddengate)
        
        return hy
```

```python
class GRUModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, layer_dim, output_dim, bias=True):
        super(GRUModel, self).__init__()
        # Hidden dimensions
        self.hidden_dim = hidden_dim
         
        # Number of hidden layers
        self.layer_dim = layer_dim
         
        self.gru_cell = GRUCell(input_dim, hidden_dim, bias=True)
        
        self.fc = nn.Linear(hidden_dim, output_dim)
     
    
    # 前向传播，x为输入张量，手写数字x.shape = (batch_size, sequence_length, input_dim)，x.shape = (100, 28, 28)
    # 表示有 100 个样本（batch）每个样本有 28 个时间步（序列长度）每个时间步是一个 28 维的向量（每一行像素）
    def forward(self, x):
        #print(x.shape,"x.shape")100, 28, 28
        h0 = torch.zeros(self.layer_dim, x.size(0), self.hidden_dim).to(device)
         
       
        outs = [] # 收集每一个时间步的输出。
        
        hn = h0[0,:,:] #张量切片 表示取出第 0 层（因为只有 1 层），留下所有 batch 和 hidden 维度，这时hn = h0[0]  # 等价于 h0[0,:,:]
        # seq当前时间步
        for seq in range(x.size(1)):
            hn = self.gru_cell(x[:,seq,:], hn) 
            outs.append(hn)
            
        out = outs[-1].squeeze() #只取最后一个时间步的隐藏层，作为输出。.squeeze() 去掉多余的维度，变成 (batch_size, hidden_dim)。
        
        out = self.fc(out) 
        # out.size() --> 100, 10
        return out
```

## LSTM cell implmentation

```python
class LSTMCell(nn.Module):
    def __init__(self,input_size,hidden_size,bias=True):
        super(LSTMCell,self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.bias = bias
        self.x2h = nn.Linear(input_size,4 * hidden_size,bias = bias)
        self.h2h = nn.Linear(hidden_size,4 * hidden_size,bias = bias)
        self.reset_parameters()
        
    def reset_parameters(self):
        std = 1.0/math.sqrt(self.hidden_size)
        for w in self.parameters():
            w.data.uniform_(-std,std)
            
    def forward(self,x,hidden):
        hx, cx = hidden
        x = x.view(-1,x.size(1))
        gates = self.x2h(x) + self.h2h(hx) # 输入维度+隐层维度
        gates = gates.squeeze() #移除大小为1的维度
        
        inputgate,forgetgate,cellgate,outputgate = gates.chunk(4,1)
        
        inputgate = torch.sigmoid(inputgate)
        forgetgate = torch.sigmoid(forgetgate)
        cellgate = torch.tanh(cellgate)
        outputgate = torch.sigmoid(outputgate)
        
        cy = torch.mul(cx,forgetgate) + torch.mul(inputgate,cellgate) # orch.mul 是逐元素乘法，适用于相同形状的张量，对应位置的元素相乘。
        hy = torch.mul(outputgate,torch.tanh(cy))
        
        return (hy,cy)
```

```python
'''
STEP 2: CREATE MODEL CLASS
'''
class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, layer_dim, output_dim, bias=True):
        super(LSTMModel, self).__init__()
        self.hidden_dim = hidden_dim
        self.layer_dim = layer_dim   
        self.lstm = LSTMCell(input_dim, hidden_dim)  
        self.fc = nn.Linear(hidden_dim, output_dim)
     
    def forward(self, x):
        
        #print(x.shape,"x.shape")100, 28, 28
        h0 = torch.zeros(self.layer_dim, x.size(0), self.hidden_dim).to(device)
        c0 = torch.zeros(self.layer_dim, x.size(0), self.hidden_dim).to(device)

        outs = []
        
        cn = c0[0,:,:]
        hn = h0[0,:,:]

        for seq in range(x.size(1)):
            hn, cn = self.lstm(x[:,seq,:], (hn,cn)) 
            outs.append(hn)

        out = outs[-1].squeeze()
        
        out = self.fc(out) 
        # out.size() --> 100, 10
        return out
```

## 训练

```python
input_dim = 28
hidden_dim = 128
layer_dim = 1  # ONLY CHANGE IS HERE FROM ONE LAYER TO TWO LAYER
output_dim = 10
 
model = LSTMModel(input_dim, hidden_dim, layer_dim, output_dim).to(device)
# model = GRUModel(input_dim, hidden_dim, layer_dim, output_dim).to(device)
criterion = nn.CrossEntropyLoss()
learning_rate = 1e-3
 
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)  
```

```python
seq_dim = 28 # Number of steps to unroll # 每个样本的时间步长度（如28x28图像按行展开成28步）

loss_list = []
iter = 0

def evaluate_accuracy(model, data_loader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in data_loader:
            images = images.view(-1, seq_dim, input_dim).to(device)
            labels = labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    accuracy = 100 * correct / total
    return accuracy

for epoch in range(num_epochs):
    model.train()
    loop = tqdm(enumerate(train_loader), total=len(train_loader), desc=f"Epoch {epoch+1}/{num_epochs}")
    # enumerate(train_loader)：遍历训练数据加载器，返回索引和数据。desc：设置进度条的描述信息。
    for i, (images, labels) in loop:
        # Load images as Variable
        images = images.view(-1, seq_dim, input_dim).to(device)
        labels = labels.to(device)

        outputs = model(images)        # Forward pass to get output/logits   前向传播     # outputs.size() --> 100, 10 
        loss = criterion(outputs, labels)# Calculate Loss: softmax --> cross entropy loss
        optimizer.zero_grad()# Clear gradients w.r.t. parameters
        loss.backward()        # Getting gradients w.r.t. parameters
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)  # 梯度裁剪
        optimizer.step()        # Updating parameters
        
        loss_list.append(loss.item())
        iter += 1
        
        loop.set_postfix(loss=loss.item())#通常是一个 tqdm 进度条对象，用于可视化训练/测试的进度。set_postfix 方法：动态更新进度条后显示的附加信息（如当前损失值、准确率等）。
        
        if iter % 600 == 0:             
            accuracy = evaluate_accuracy(model, test_loader)
            loop.write('Iteration: {}. Loss: {}. Accuracy: {}'.format(iter, loss.item(), accuracy))  # 使用 loop.write 打印
print("Training complete.")
```
